```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
#Extract first CSV into DataFrame
csv_file = "Datafiniti_Hotel_Reviews.csv"
datafiniti_df = pd.read_csv(csv_file)
datafiniti_df.head()
```

```python
#Extract second CSV into DataFrame
csv_file_two = "Datafiniti_Hotel_Reviews_Jun19.csv"
jun19_datafiniti_df = pd.read_csv(csv_file_two)
jun19_datafiniti_df.head()
```

```python
# Create a filtered dataframe from specific columns
datafiniti_cols = ['id', 'name', 'address', 'latitude', 'longitude', 'postalCode', 'city', 'province', 'country', 'reviews.username', 'reviews.userCity', 'reviews.userProvince', 'reviews.date', 'reviews.rating']
datafiniti_transformed= datafiniti_df[datafiniti_cols].copy()

# Rename the column headers
datafiniti_transformed = datafiniti_transformed.rename(columns={"id": "hotel_id", "postalCode": "zipcode", "province": "state", "reviews.username": "username", "reviews.userCity": "user_city", "reviews.userProvince": "user_state", "reviews.date": "date", "reviews.rating": "rating"})

# Clean the data by dropping duplicates and setting the index
datafiniti_transformed.drop_duplicates("hotel_id", inplace=True)
datafiniti_transformed.set_index

#("id", inplace=True)

datafiniti_transformed.head()
```

```python
datafiniti_transformed.dtypes
```

```python
datafiniti_transformed.count
```

```python
# Create a filtered dataframe from specific columns
jun19_datafiniti_cols = ['id', 'name', 'address', 'latitude', 'longitude', 'postalCode', 'city', 'province', 'country', 'reviews.username', 'reviews.userCity', 'reviews.userProvince', 'reviews.date', 'reviews.rating']
jun19_datafiniti_transformed= jun19_datafiniti_df[jun19_datafiniti_cols].copy()

# Rename the column headers
jun19_datafiniti_transformed = jun19_datafiniti_transformed.rename(columns={"id": "hotel_id", "postalCode": "zipcode", "province": "state", "reviews.username": "username", "reviews.userCity": "user_city", "reviews.userProvince": "user_state", "reviews.date": "date", "reviews.rating": "rating"})

# Clean the data by dropping duplicates and setting the index
jun19_datafiniti_transformed.drop_duplicates("hotel_id", inplace=True)
jun19_datafiniti_transformed.set_index
#("hotel id", inplace=True)

jun19_datafiniti_transformed.head()
```

```python
jun19_datafiniti_transformed.dtypes
```

```python
jun19_datafiniti_transformed.count
```

```python
#Combine both cleaned dataframes and drop duplicates 
combined_transformed_df = pd.concat([datafiniti_transformed,jun19_datafiniti_transformed]).drop_duplicates()
```

```python
combined_transformed_df.head()
```

```python
combined_transformed_df.count
```

```python
#Create new data with select columns
hotels_df = combined_transformed_df[['hotel_id', 'name', 'address', 'latitude', 'longitude', 'zipcode', 'city', 'state', 'country' ]].copy()
hotels_df.set_index('hotel_id', inplace=True)
hotels_df.head()
```

```python
#Create new data with select columns
users_df = combined_transformed_df[['username', 'user_city', 'user_state']].copy()
users_df.set_index('username', inplace=True)
users_df.head()
```

```python
#Create new data with select columns
reviews_df = combined_transformed_df[['hotel_id', 'username', 'date', 'rating']].copy()
reviews_df.reset_index()
reviews_df.head()
```

```python
#Create database connection
connection_string = "postgres:postgres@localhost:5432/Hotel_Reviews"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
#Load DataFrames into database
hotels_df.to_sql(name='hotels', con=engine, if_exists='append', index=True)
```

```python
users_df.to_sql(name='users', con=engine, if_exists='append', index=True)
```

```python
reviews_df.to_sql(name='reviews', con=engine, if_exists='append', index=False)
```
